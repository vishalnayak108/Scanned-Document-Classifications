## Convert images to text

```python
def images_to_txt(path, language='eng'):
    images = pdf2image.convert_from_path(path, first_page=0, last_page=2)
    all_text = []
    for page in images:
        pil_im = page
        try:
          # Detect orientation and rotate the image if necessary
          osd = pytesseract.image_to_osd(page, lang='eng', config='--psm 0')
          rotation_angle = float(osd.split("\n")[0].split(":")[1])
          page = page.rotate(rotation_angle, expand=True)
        except:
          pass
        text = pytesseract.image_to_string(page, lang=language)
        # ocr_dict = pytesseract.image_to_data(pil_im, lang='eng', output_type=Output.DICT)
        # ocr_dict now holds all the OCR info including text and location on the image
        # text = " ".join(ocr_dict['text'])
        # text = re.sub('[ ]{2,}', '\n', text)
        all_text.append(text)
    final_text= " ".join(all_text)
    return final_text, len(all_text)
```

```python
import glob
import pandas as pd

root_dir = './dataset'
df = pd.DataFrame(columns=['text', 'label'])

for dir in os.listdir(root_dir):
    label = dir
    print(label)
    for pdf_file in glob.glob(f"{root_dir}/{dir}/*.pdf"):
        print(pdf_file)
        all_text, len_all_text = images_to_txt(pdf_file)
        df = df.append({'text': str(all_text), 'label': label},ignore_index=True)
```

### Label Encoding

```python
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()
```

```python
# encoding train labels
encoder.fit(data['label'])
y = encoder.transform(data['label'])
```

```python
print(encoder.classes_)
```

### Text preprocessing

```python
import string

punct =[]
punct += list(string.punctuation)
punct += '’'
punct.remove("'")
def remove_punctuations(text):
    # print(text)
    for punctuation in punct:
        text = text.replace(punctuation, ' ')
    return text
```

```python
import nltk
nltk.download('stopwords')
#Stop words present in the library
stopwords = nltk.corpus.stopwords.words('english')
#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    tokens = text.split()
    output= " ".join([i for i in tokens if i not in stopwords])
    # print(output)
    return output
```

```python
data['text'] = data['text'].apply(lambda x: remove_punctuations(x))
data['text'] = data['text'].apply(lambda x: str(x).replace("  ", " "))
data['text'] = data['text'].apply(lambda x: remove_stopwords(x))
data['text'] = data['text'].apply(lambda x: x.lower())
```

### Train-Test split (70:30 ratio)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data['text'], y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import LinearSVC
import numpy as np
from sklearn.metrics import accuracy_score, classification_report
```

### Logistic model

```python
model_logistic = Pipeline([('vect', CountVectorizer(min_df=5, ngram_range=(1,2))),
                      ('tfidf', TfidfTransformer()),
                      ('model',LogisticRegression()),
                     ])

model_logistic.fit(X_train, y_train)

ytest = np.array(y_test)
pred = model_logistic.predict(X_test)
```

```python
print('accuracy %s' % accuracy_score(pred, y_test))
print(classification_report(ytest, pred))
```

SVM Model

```python
svc = Pipeline([('vect', CountVectorizer(min_df=5, ngram_range=(1,2))),
               ('tfidf', TfidfTransformer()),
               ('model',LinearSVC()),
               ])

svc.fit(X_train, y_train)

ytest = np.array(y_test)
y_pred = svc.predict(X_test)
```

```python
print('accuracy %s' % accuracy_score(y_pred, y_test))
print(classification_report(ytest, y_pred))
```

### Naive Bayes Classifier

```python
nbc = Pipeline([('vect', CountVectorizer(min_df=5, ngram_range=(1,2))),
               ('tfidf', TfidfTransformer()),
               ('model',MultinomialNB()),
               ])

nbc.fit(X_train, y_train)

ytest = np.array(y_test)
pred_y = nbc.predict(X_test)
```

```python
print('accuracy %s' % accuracy_score(pred_y, y_test))
print(classification_report(ytest, pred_y))
```

### Random Forest

```python
rf = Pipeline([('vect', CountVectorizer(min_df=5, ngram_range=(1,2))),
               ('tfidf', TfidfTransformer()),
               ('rf', RandomForestClassifier(n_estimators=50)),
               ])

rf.fit(X_train, y_train)

ytest = np.array(y_test)
preds = rf.predict(X_test)
```

```python
print('accuracy %s' % accuracy_score(preds, y_test))
print(classification_report(ytest, preds))
```

###Dataset Description:

The dataset used for this classification task was relatively small.
It was observed that the dataset was highly unbalanced, with some classes having significantly fewer samples compared to others.
###Data Preparation and Pre-processing:

The project began with data engineering and text pre-processing. This included steps like removing punctuation, eliminating stop words, and performing stemming operations to clean and prepare the text data for modeling.
The decision was made to focus on the first 2-3 pages of text content for classification, as this was deemed sufficient to determine the document type. Analyzing the entire document text was avoided to reduce processing time and model latency.
###Feature Engineering:

The text data was converted into numerical feature vectors using a Bag of Words (BoW) model. BoW is a traditional method for text representation in which the frequency of each word is counted, and these counts are used as features for classification.
###Model Selection:

Various classical machine learning models were considered and implemented, including Logistic Regression, Support Vector Machines (SVM), and Random Forest.
The decision was made not to employ more complex and computationally expensive models such as neural networks, Convolutional Neural Networks (CNNs), or BERT-based models. This choice was driven by the desire to keep model latency low, given the project's focus on processing only the first 2-3 pages of text.
###Model Evaluation:

Given the imbalance in the dataset, the evaluation metrics of recall, precision, and F1 score were chosen as the primary measures of model performance.
These metrics are well-suited for imbalanced datasets, as they provide insights into the model's ability to correctly classify minority classes while maintaining a good balance between precision and recall.
###Results:

The classical machine learning models (Logistic Regression, SVM, Random Forest) yielded pretty decent results.
By optimizing these models and fine-tuning hyperparameters, satisfactory classification performance was achieved without the need for more complex models.
The decision to focus on the initial pages of text content and use text pre-processing techniques helped streamline the model's efficiency and maintain acceptable latency.

In summary, this approach leveraged classical machine learning methods, careful data preparation, and feature engineering to achieve good results in a multiclass classification task with a small and unbalanced dataset, while also keeping model latency in check. The choice of appropriate evaluation metrics ensured that the model's performance was assessed effectively, considering the dataset's characteristics.
